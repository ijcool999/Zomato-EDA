```python
# Extracting data
X = train_data["Average Cost for two"]
Y = train_data["Aggregate rating"]

# Computing the least squares estimates
n = len(X)
w1 = (np.sum(Y*X) - (1/n)*np.sum(X)*np.sum(Y)) / (np.sum(X*X) - (1/n)*np.sum(X)*np.sum(X))
w0 = (1/n) * np.sum(Y) - w1 * (1/n) * np.sum(X)

print("Intercept (w0):", w0)
print("Slope (w1):", w1)
```

```python
Y_pred = w0 + w1 * X

# Computing the squared errors
squared_errors = (Y - Y_pred) ** 2

# Computing the Average Squared Error (ASE)
ASE = np.mean(squared_errors)
print("Average Squared Error (ASE):", ASE)
```

```python
import numpy as np

# Extracting variables from the train_data dataframe
x = np.asanyarray(train_data["Average Cost for two"])
y = np.asanyarray(train_data["Aggregate rating"])

# Initial values for w0 and w1
w0 = [1]  
w1 = [1]  

# Learning rate
eta = 0.1e-10

# Number of iterations
num_iterations = 16

# List to store the cost at each iteration
cost = []  

# Gradient descent algorithm
for i in range(num_iterations):
    # Predicted values
    predicted = w0[i] + w1[i] * x
    
    # Absolute error
    error = np.abs(predicted - y)
    
    # Cost
    total_cost = np.sum(error)
    cost.append(total_cost)
    print("Iteration:", i, "Cost:", total_cost)
    
    # Partial derivatives
    dw0 = -2 * np.sum(error)  # Partial derivative of cost w.r.t. w0
    dw1 = -2 * np.sum(error * x)  # Partial derivative of cost w.r.t. w1
    
    # Update rule for w0 and w1
    w0.append(w0[i] - eta * dw0)
    w1.append(w1[i] - eta * dw1)

# Print the final values of w0 and w1
print("Final values:")
print("w0:", w0[-1])
print("w1:", w1[-1])
```

```python
# probability of a restaurant receiving an "Excellent" rating given certain conditions such as having table booking, online delivery, etc.
import pandas as pd

# Load the dataset
df = pd.read_csv('zomato.csv')

# Selecting columns
col = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Price range', 'Aggregate rating', 'Votes']
df1 = df[col]

# Calculating probabilities
p_excellent = (df1['Aggregate rating'] >= 4.0).sum() / len(df1)

p_has_table_booking = (df1['Has Table booking'] == 'Yes').sum() / len(df1)
p_has_online_delivery = (df1['Has Online delivery'] == 'Yes').sum() / len(df1)
p_is_delivering_now = (df1['Is delivering now'] == 'Yes').sum() / len(df1)

p_price_range_1 = (df1['Price range'] == 1).sum() / len(df1)
p_price_range_2 = (df1['Price range'] == 2).sum() / len(df1)
p_price_range_3 = (df1['Price range'] == 3).sum() / len(df1)
p_price_range_4 = (df1['Price range'] == 4).sum() / len(df1)

# Conditional probabilities
p_excellent_given_conditions = p_excellent * p_has_table_booking * p_has_online_delivery * p_is_delivering_now * p_price_range_1

# Printing the result
print("Probability of getting an 'Excellent' rating given the conditions:", p_excellent_given_conditions)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('zomato.csv')

# Calculate the total number of restaurants
total_restaurants = len(df)

# Define the cuisines to analyze
cuisines = ['Italian', 'Indian', 'Chinese', 'Continental']

# Calculate the probabilities for each cuisine
cuisine_probabilities = {}
for cuisine in cuisines:
    cuisine_count = len(df[df['Cuisines'].str.contains(cuisine, case=False, na=False)])
    cuisine_probabilities[cuisine] = cuisine_count / total_restaurants

print("Probabilities for each cuisine:")
for cuisine, probability in cuisine_probabilities.items():
    print(f"Probability of offering {cuisine} cuisine: {probability:.6f}")
```

```python
# Define the rating threshold
rating_threshold = 4.0

# Calculate the probability of having a high rating
high_rating_count = len(df[df['Aggregate rating'] > rating_threshold])
high_rating_probability = high_rating_count / total_restaurants

print(f"\nProbability of having a rating above {rating_threshold} stars: {high_rating_probability:.6f}")
```

```python
# Calculate the probability of offering online delivery
online_delivery_count = len(df[df['Has Online delivery'] == 'Yes'])
online_delivery_probability = online_delivery_count / total_restaurants

print(f"\nProbability of offering online delivery: {online_delivery_probability:.6f}")
```
